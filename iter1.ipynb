```python
import pandas as pd
import tensorflow as tf
from nba_api.stats.static import teams
from nba_api.stats.endpoints import leaguegamefinder
```

```python
nba_teams = teams.get_teams()
# nba_abbrev = ['ATL', 'BOS', 'BKN', 'CHA', 'CHI', 'CLE', 'DAL', 'DEN', 'DET', 'GSW', 'HOU', 'IND', 'LAC', 'LAL', 'MEM', 'MIA', 'MIL', 'MIN', 'NOP', 'NYK', 'OKC', 'ORL', 'PHI', 'PHX', 'POR', 'SAC', 'SAS', 'TOR', 'UTA', 'WAS']
# nba_teams = [team for team in nba_teams if team['abbreviation'] in nba_abbrev]
# nba_teams_id = [i['id'] for i in nba_teams]
# seasons = ['2018-19', '2019-20']
gamefinder = leaguegamefinder.LeagueGameFinder(league_id_nullable='00')
games = gamefinder.get_data_frames()[0]
games.head()
```

```python
games.dtypes
```

```python
obj_columns = ['TEAM_ABBREVIATION', 'TEAM_NAME', 'GAME_DATE', 'MATCHUP', 'WL']
for i in obj_columns:
    games[i] = pd.Categorical(games[i])
    games[i] = games[i].cat.codes 
```

```python
games.head()
```

```python
target = games.pop('WL')
games.pop('SEASON_ID')
games.pop('TEAM_ID')
games.pop('GAME_ID')
```

```python
games.head()
```

```python
dataset = tf.data.Dataset.from_tensor_slices((games.values, target.values))
```

```python
games.head()
```

```python
for feat, targ in dataset.take(5):
  print ('Features: {}, Target: {}'.format(feat, targ))
```

```python
train_dataset = dataset.shuffle(len(games)).batch(1)
```

```python
def get_compiled_model():
  model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(1)
  ])

  model.compile(optimizer='adam',
                loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
                metrics=['accuracy'])
  return model
```

```python
model = get_compiled_model()
model.fit(train_dataset, epochs=15)
```
